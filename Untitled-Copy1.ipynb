```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',style='time')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',style='time',kind='line')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',col='time')
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tips)
```

```python
sns.catplot(x='total_bill',y='day',kind='violin',data=tips,hue='sex',split='True')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
sns.swarmplot(x='day',y='total_bill',data=tips,color='k',size=3)
```
